# House Price Index Data Clean

```python
from os import listdir
import pandas as pd

import pandas as pd
pd.set_option('display.height', 1000)
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns',  500)
pd.set_option('display.width', 1000)
```

```python
FROMPATH = 'data/raw/'
TOPATH = 'data/clean/'
listdir(FROMPATH)
```

### Price Per Region

```python
# Not awful data
ppr = pd.read_csv(FROMPATH + 'pricePerRegion.csv', encoding='utf-8')
ppr[:5]
```

```python
# Set index
ppr['Date'] = ppr['Unnamed: 0']
ppr = ppr.set_index('Date').drop('Unnamed: 0', axis=1)
ppr = ppr.iloc[1:]
ppr = ppr.rename(str.lower, axis='columns')
ppr[:5]
```

```python
# Getting rid of unneeded fields
toDrop = ['Unnamed: 37', 'Unnamed: 34', 'NORTH EAST', 'NORTH WEST',
           'YORKS & THE HUMBER', 'EAST MIDLANDS', 'WEST MIDLANDS', 'EAST OF ENGLAND',
           'LONDON', 'SOUTH EAST', 'SOUTH WEST', 'England', 'Unnamed: 49', 'Unnamed: 50', 
            'Unnamed: 47']

for field in toDrop:
    ppr = ppr.drop(field.lower(), axis=1)
ppr[:5]
```

```python
ppr.to_csv(TOPATH + 'pricePerRegion.csv', index=True)
```

## Price Per Type

```python
# Awful data...
ppt = pd.read_csv(FROMPATH + 'pricePerType.csv')
ppt[:5]
```

```python
toDrop = ['UNITED KINGDOM'] + ['Unnamed: %s' % i for i in range(6, 18)]
for field in toDrop:
    if field in ppt.keys():
        ppt = ppt.drop(field, axis=1)
ppt[:5]
```

```python
from time import strptime

month = []
for i in range(len(ppt)):
    point = ppt['Unnamed: 1'].iloc[i]
    if str(point) != 'nan':
        try:
            month.append(strptime(point, '%b').tm_mon)
        except ValueError:
            month.append(0)
    else:
        month.append(0)
ppt['month'] = month
ppt = ppt.drop('Unnamed: 1', axis=1)
ppt['month'].iloc[1] = 'Month'
ppt[:5]
```

```python
tmpYear = 0
year = []
for i in range(len(ppt)):
    if str(ppt['Unnamed: 0'].iloc[i]) != 'nan':
        tmpYear = ppt['Unnamed: 0'].iloc[i]
    year.append(int(tmpYear))
ppt['Unnamed: 0'] = year
ppt['Unnamed: 0'].iloc[1] = 'Year'
ppt[:5]
```

```python
ppt.columns = ppt.iloc[1]
ppt.drop
ppt = ppt[2:]
ppt[:5]
```

```python
ppt.to_csv(TOPATH + 'pricePerType.csv', index=False)
```

## Sales Volume

```python
sV = pd.read_csv(FROMPATH + 'salesVolume.csv')
sV[:5]
```

```python
toDrop = ['Unnamed: 37', 'Unnamed: 34', 'NORTH EAST', 'NORTH WEST',
           'YORKS & THE HUMBER', 'EAST MIDLANDS', 'WEST MIDLANDS', 'EAST OF ENGLAND',
           'LONDON', 'SOUTH EAST', 'SOUTH WEST', 'England', 'Unnamed: 47'] \
        + ['Unnamed: %s' % i for i in range(49, 255)]

for key in toDrop:
    if key in sV.keys():
        sV = sV.drop(key, axis=1)
sV = sV[1:275]
sV[:5]
```

```python
month = []
year = []

for i in range(len(sV)):
    point = sV['Unnamed: 0'].iloc[i].split('-')
    if int(point[1]) > 50:
        year.append(int('19' + point[1]))
    else:
        year.append(int('20' + point[1]))
    month.append(strptime(point[0], '%b').tm_mon)
    
sV['year'] = year
sV['month'] = month
sV = sV.drop('Unnamed: 0', axis=1)
sV[:5]
```

```python
sV = sV.rename(str.lower, axis='columns')
sV[:2]
```

```python
sV.to_csv(TOPATH + 'saleVolume.csv', index=False)
```
