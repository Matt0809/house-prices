```python
uQ = t[5]
lQ = t[1]
lQ['Area'].iloc[0] 
uQL = []
lQL = []
fails = []
for i in range(len(t[0])):
    row = t[0].iloc[i]
    district = row['District'] if row['District'] != 'city of westminster' else 'westminster'
    district = district if district != 'epping forest' else 'outer london'
    ind = uQ.index[uQ['Area'] == district]
    if len(ind) != 0:
        try:
            year = row['yearOfTransfer'] - 1 if (row['yearOfTransfer'] -1 ) > 1995 else 1996
            uQL.append(uQ.iloc[ind[0]][str(year)])
            lQL.append(lQ.iloc[ind[0]][str(year)])
        except KeyError:
            print('Failed at %s' % row['yearOfTransfer'])
            uQL.append(np.nan)
            lQL.append(np.nan)
    else:
        uQL.append(np.nan)
        lQL.append(np.nan)
        fails.append(district)
```

```python
t[0]['lowerQuartile'] = lQL
t[0]['upperQuartile'] = uQL
```

```python
# lack of index messup. Fixed on DataClean2 but have to load indiv
ppR = pd.read_csv(DATAPATH + 'pricePerRegion.csv')
ppR[:2]
# not very clean...
```

```python
# Cleaner
from time import strptime

monthList = []
yearList = []

for i in range(len(ppR)):
    try:
        point = ppR['Date'].iloc[i].split('-')
        monthList.append(int(strptime(point[0], '%b').tm_mon))
        year = int('20' + point[1]) if int(point[1]) < 50 else int('19' + point[1])
        yearList.append(int(year))
    except AttributeError:
        monthList.append(np.nan)
        yearList.append(np.nan)
        
ppR['year'] = yearList
ppR['month'] = monthList

ppR = ppR[:-1].drop('Date', axis=1)
ppR[:5]
```

```python
# Same format so might as well
sV = pd.read_csv(DATAPATH + 'saleVolume.csv')
sV[:5]
```

```python
ppRL = []
sVL = []
n = 0

for i in range(len(t[0])):
    row = t[0].iloc[i]
    year = row['yearOfTransfer']
    month = row['monthOfTransfer']
    district = row['District'] if row['District'] != 'city of westminster' else 'westminster'
    district = district if district != 'epping forest' else 'outer london'
    district = district if district != 'kensington and chelsea' else 'kensington & chelsea'
    district = district if district != 'hammersmith and fulham' else 'hammersmith & fulham'
    indR = ppR.index[[ppR['year'].iloc[i] == year and ppR['month'].iloc[i] == month 
                    for i in range(len(ppR))]]
    indSV = sV.index[[sV['year'].iloc[i] == year and sV['month'].iloc[i] == month 
                    for i in range(len(sV))]]
    try:
        ppRL.append(ppR.iloc[indR][district])
        sVL.append(sV.iloc[indSV][district])
    except KeyError:
        n += 1
        ppRL.append(ppR.iloc[indR]['inner london'])
        sVL.append(sV.iloc[indSV]['inner london'])
        
print('Number of nan rows: %s' % n)
```

```python
ppTList = []
ppRList = []
sVList = []
n = 0

def removeExcess(point):
    try:
        value = float((point.split('\n')[0]).split('   ')[1])
    except IndexError:
        value = np.nan
    except AttributeError:
        value = np.nan
    return value

for i in range(len(joined)):
    ppTList.append(removeExcess(joined['pricePerType'].iloc[i]))
    ppRList.append(removeExcess(joined['pricePerRegion'].iloc[i]))
    sVList.append(removeExcess(joined['salesVolume'].iloc[i]))
    if i % len(joined)/100 == 0:
        if n < 10:
            print('\b\b' + str(n)+'%')
        else:
            print('\b\b\b' + str(n) + '%')
        n += 1
```

```python
joined['salesVolume'] = sVList
joined['pricePerType'] = ppTList
joined['pricePerRegion'] = ppRList
```
